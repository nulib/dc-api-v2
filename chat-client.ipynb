```python
import websocket
import json

class NulDCChatClient:
  def __init__(self, addr, auth_token, on_token=None, on_answer=None, on_error=None, on_debug=None):
    self.addr = addr
    self.auth_token = auth_token
    self.handlers = {
      'token': on_token,
      'answer': on_answer,
      'error': on_error,
      'debug': on_debug
    }

  def __dispatch(self, key, data):
    if self.handlers[key]:
      self.handlers[key](data)

  def ask(self, q, ref=''):
    answer = None
    def on_message(app, data):
      nonlocal answer
      message = json.loads(data)
      if 'token' in message:
        self.__dispatch('token', message['token'])
      if 'answer' in message:
        answer = message['answer']
        self.dispatch('answer', answer)
        app.close()
      if message.get('type', None) == 'debug':
        self.dispatch('debug', message['message'])
      if 'error' in message:
        self.dispatch('error', message['error'])
        app.close()

    payload = { 'message': 'chat', 'auth': self.auth_token, 'ref': ref, 'question': q }
    ws = websocket.WebSocketApp(addr, on_message=on_message, on_open=lambda ws: ws.send(json.dumps(payload)))
    ws.run_forever()
    return answer
```

```python
client = NulDCChatClient(addr, auth_token=token, on_token=lambda t: print(t))
answer = client.ask(question)
```
